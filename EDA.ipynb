```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv(r'C:\Project\myProject\Prognozowanie-rezygnacji-klientw-dla-firmy-telekomunikacyjnej\internet_service_churn.csv')

# Basic info and checking for missing values
print(df.info())
print(df.isnull().sum())

# Visualizing distributions
sns.histplot(df['subscription_age'], kde=True)
plt.title('Subscription Age Distribution')
plt.show()

# Correlation matrix
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
